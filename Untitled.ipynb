## EDA

```python
categorical = train.select_dtypes(include=[np.object])
categorical = categorical.columns

numeric = train.select_dtypes(include=[np.int,np.float])
numeric = numeric.columns
```

### Categorical

```python
count = 0
# plt.figure(figsize=(12,12))
done_list = []
plt.figure(figsize=(12,4))
for i in categorical.drop(['var30']):
    count+=1
    if count<5:
#         plt.subplot(int(str()+str(count)))
        ## fill your code here for plotting the countplot
        sns.countplot(train[i])
        done_list.append(i)
        plt.title(i)
        plt.show()
```

```python
train['var34'].value_counts()
```

```python
#         plt.subplot(int(str()+str(count)))
        ## fill your code here for plotting the countplot
var = pd.crosstab(train['var34'],train['self_service_platform'])
print(var.sum(1).astype(float))
print(var)
```

```python
count = 0
# plt.figure(figsize=(12,12))
plt.figure(figsize=(8,4))
done_list_norm = []
for i in categorical.drop(['var30']):
    count+=1
    if count<5:
#         plt.subplot(int(str()+str(count)))
        ## fill your code here for plotting the countplot
        var = pd.crosstab(train[i],train['self_service_platform']) 
        var.div(var.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(20,5))
        done_list_norm.append(i)
        plt.title(i)
        plt.show()
```

### Continuous Variables

```python
count = 0
done_list_num  = []

plt.figure(figsize=(12,12))

for i in numeric.drop(['cust_id']):
    if count<9:
        count+=1
        plt.subplot(int(str(33)+str(count)))
        
        ### fill your code here to plot the kdeplot
        sns.kdeplot(train[i])

        plt.title(i)
        
        done_list_num.append(i)
        
plt.show()
```
